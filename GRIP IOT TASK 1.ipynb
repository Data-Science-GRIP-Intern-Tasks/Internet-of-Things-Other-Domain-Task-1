<h4>Apply canny blur to recognize text in the picture</h4>

```python
imgThreshold = cv.Canny(gray,100,0)

backup = imgThreshold.copy()

plt.imshow(imgThreshold)
```

<h4>Perform fundamental morphological operations</h4>

```python
kernel = np.ones((5, 5))

#To apply dilation 
imgDial = cv.dilate(imgThreshold, kernel, iterations=2) 

#To apply erosion 
imgThreshold = cv.erode(imgDial, kernel, iterations=1)
```

<h4>Detect contours</h4>

```python
#To copy image for display
imgContours = image.copy() 

imgBigContour = image.copy() 

#To find all contours
contours, hierarchy = cv.findContours(imgThreshold, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE) 

#To draw all detected contours
cv.drawContours(imgContours, contours, -1, (0, 255, 0), 10)
```
